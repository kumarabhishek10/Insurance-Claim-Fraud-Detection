### Checking the outliers

```python
# lets check the outliers using boxplot
plt.figure(figsize=(25,25),facecolor='white')
plotnumber=1
for col in numerical:
    if plotnumber<=36:
        ax=plt.subplot(12,2,plotnumber)     
        sns.boxplot(df[col],palette='bright')
        plt.xlabel(col,fontsize=15)
        plt.yticks(rotation=0,fontsize=10)
    plotnumber+=1
plt.tight_layout()
```

```python
# Using Z square mode
```

```python
from scipy.stats import zscore
out_features=df[['policy_annual_premium', 'total_claim_amount','property_claim','age']]
z=np.abs(zscore(out_features))  
z
```

```python
#threshold=3

np.where(z>3)
```

```python
# now removing thedata above zscore and dreating a new DF
df1=df[(z<3).all(axis=1)]
df1.shape
```

```python
#shape of old and new data frame
print('Old Data Frame -',df.shape[0])
print('New Data Frame -',df1.shape[0])
```

```python
# Now we will use IQR method


#1st quar=ntile
q1=out_features.quantile(0.25)
 

#3rd quantile
q3=out_features.quantile(0.75)

#IQR
IQR=q3-q1
```

```python
df2=df[~((df < (q1-1.5*IQR))|(df>(q3+1.5*IQR))).any(axis=1)] 
```

```python
df2.shape
```

```python
#shape of old and new data frame
print('Old Data Frame -',df.shape[0])
print('New Data Frame -',df2.shape[0])
```

```python
# We can see that Zquare is more effective as loss of data is less
```

#### Modelling

```python
#Importing necessary libraries



from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=i)
    RFR=RandomForestClassifier()
    RFR.fit(x_train,y_train)
    pred=RFR.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is",maxAccu,"at random_state",maxRS)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=maxRS)
```

#### Classification Algorithms

```python
#Importing all the classification libraries

from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
#Checking accuracy for Random Forest Classifier

RFC=RandomForestClassifier()
RFC.fit(x_train,y_train)
predRFC=RFC.predict(x_test)
print(accuracy_score(y_test,predRFC))
print(confusion_matrix(y_test,predRFC))
print(classification_report(y_test,predRFC))
```

```python
# accuracy is 81.1%
```

```python
#Checking accuracy for logistic regression model

LR=LogisticRegression()
LR.fit(x_train,y_train)
predLR=LR.predict(x_test)
print(accuracy_score(y_test,predLR))
print(confusion_matrix(y_test,predLR))
```

```python
# accuracy is 81.2%
```

```python
#Checking accuracy for support vector machine classifier

svc=SVC()
svc.fit(x_train,y_train)
predsvc=svc.predict(x_test)
print(accuracy_score(y_test,predsvc))
print(confusion_matrix(y_test,predsvc))
print(classification_report(y_test,predsvc))
```

```python
# accuracy is 80%
```

```python
#Checking accuracy for Gradient Boosting Classifier

GB=GradientBoostingClassifier()
GB.fit(x_train,y_train)
predGB=GB.predict(x_test)
print(accuracy_score(y_test,predGB))
print(confusion_matrix(y_test,predGB))
print(classification_report(y_test,predGB))
```

```python
# accuracy is 84.9%
```

```python
#Checking accuracy for Ada Boost Classifier

ABC=AdaBoostClassifier()
ABC.fit(x_train,y_train)
predABC=ABC.predict(x_test)
print(accuracy_score(y_test,predABC))
print(confusion_matrix(y_test,predABC))
print(classification_report(y_test,predABC))
```

```python
# accuracy is 84.2%
```

```python
#Checking accuracy for Bagging Classifier

BC=BaggingClassifier()
BC.fit(x_train,y_train)
predBC=BC.predict(x_test)
print(accuracy_score(y_test,predBC))
print(confusion_matrix(y_test,predBC))
print(classification_report(y_test,predBC))
```

```python
# accuracy is 85.2%
```

```python
#Checking accuracy for ExtraTrees Classifier

ET=ExtraTreesClassifier()
ET.fit(x_train,y_train)
predET=ET.predict(x_test)
print(accuracy_score(y_test,predET))
print(confusion_matrix(y_test,predET))
print(classification_report(y_test,predET))
```

```python
# accuracy is 80%
```
